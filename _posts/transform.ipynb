```python
import pandas as pd
import numpy as np

PATH = r'C:\Users\Hege\Downloads\submissions_20231130.csv'
df = pd.read_csv(PATH)
```

```python
df
```

```python
emails = df['email'].values
def get_domain(email):
    # if not string make it string
    if type(email) != str:
        email = str(email)
    if '@' in email:
        domain = email.split('@')[1]
        # lower case
        domain = domain.lower()
        if '.' in domain:
            domain = '.'.join(domain.split('.')[-2:])
            return domain
email_domains = [get_domain(email) for email in emails]
email_domains = list(set(email_domains))
```

```python
institution_dict = {
    'lmu.de': 'Ludwig-Maximilians-Universität München',
    'tum.de': 'Technische Universität München',
    'ifo.de': 'ifo Institut',
    'uni-konstanz.de': 'Universität Konstanz',
    'bmw.de': 'BMW',
    'ai-agency.de': 'Baiosphere',
    'acatech.de': 'Acatech',
    'umcutrecht.nl': 'UMC Utrecht',
    'rwth-aachen.de': 'RWTH Aachen',
    'helmholtz-munich.de': 'Helmholtz Zentrum München',
    'mpg.de': 'Max-Planck-Gesellschaft',
    'ku.de': 'Katholische Universität Eichstätt-Ingolstadt',
}
# Deduplicate by name
df = df.drop_duplicates(subset=['name'])
# Give each person their institution
df['institution'] = df['email'].apply(lambda x: institution_dict.get(get_domain(x), np.nan))
# Drop people without institution
df_save = df.dropna(subset=['institution'])
# Save name and institution
save_path = r'C:\Users\Hege\Downloads\TeilnehmerSymposium.xlsx'
df_save[['name', 'institution']].to_excel(save_path, index=False)
```
